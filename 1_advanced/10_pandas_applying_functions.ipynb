```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# loading data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# data cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_salaries = df[df['salary_year_avg'] > 0].copy()
df_salaries['salary_year_avg'] = df_salaries['salary_year_avg'].astype(int) 

df_salaries
```

```python
def projected_salary(salary):
  return salary * 1.03

df_salaries['salary_year_avg'] = df_salaries['salary_year_avg'].apply(projected_salary)

df_salaries
```
