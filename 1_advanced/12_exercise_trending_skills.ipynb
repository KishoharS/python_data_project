```python
#importing libraries
import ast
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# loading data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()

# data cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst']

df_DA
```

```python
df_DA['job_posted_month_no'] = df['job_posted_date'].dt.month.copy()

df_DA
```

```python
df_skills_exploded = df_DA.explode('job_skills')

df_skills_exploded
```

```python
df_DA_pivot = df_skills_exploded.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_pivot
```

```python
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()

df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_pivot = df_DA_pivot.drop('Total')

df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :5].plot(kind='line')

plt.title('Trending Skills for Data Analyst Jobs')
plt.xlabel('')
plt.ylabel('Number of Job Postings')
plt.show()
```
